```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
file_path = 'E:\\VIT_SEM2_Projects\\Bigdata_DA\\emails.csv'
df = pd.read_csv(file_path)

# Display the first few rows to understand the data
print(df.head())
```

```python
# Basic information about the dataset
print(df.info())

# Check for missing values
missing_values = df.isnull().sum()
print("Missing values in each column:\n", missing_values)
```

```python
# If there are missing values, you might want to remove them or impute them
# For now, let's remove any rows with missing values
df_cleaned = df.dropna()

# Check how many rows were removed
print("Rows before cleaning:", len(df))
print("Rows after cleaning:", len(df_cleaned))
```

```python
# Optionally, visualize the distribution of email lengths to understand the data better
# Assuming 'message' column contains the email body
df_cleaned['email_length'] = df_cleaned['message'].apply(len)
df_cleaned['email_length'].hist(bins=50)
plt.title('Distribution of Email Lengths')
plt.xlabel('Length of email body')
plt.ylabel('Frequency')
plt.show()
```
